```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('/home/danial/Data Science/Credit Risk Analysis/data/feature-engineered/credit_risk_feature_engineered.csv')
df.head()
```

```python
df = df.drop(columns = ['ID'])
```

```python
df['EDUCATION'] = df['EDUCATION'].replace({0:4 , 5:4 , 6:4})
df['MARRIAGE'] = df['MARRIAGE'].replace({0:3})
```

```python
df = pd.get_dummies(df , columns = ['SEX' , 'EDUCATION' , 'MARRIAGE'] , drop_first = True)
```

```python
target = 'default.payment.next.month'
x = df.drop(columns = [target])
y = df[target]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)
```

```python
X_scaled = pd.DataFrame(X_scaled , columns = x.columns)
final_df = pd.concat([X_scaled , y.reset_index(drop = True)] , axis = 1)
final_df.to_csv('/home/danial/Data Science/Credit Risk Analysis/data/processed/credit_ready.csv')
```
